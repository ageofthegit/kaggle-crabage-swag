```
import pandas as pd
data = pd.read_csv('kaggle-crabage-swag-train.csv')
data.head()
```

```
data.describe()
```

```
import matplotlib.pyplot as plt
plt.hist(data['Age'], bins=30, alpha=0.5, color='g')
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Histogram of Crab Age')
plt.grid(True)
plt.show()
```

```
data[['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight', 'Viscera Weight', 'Shell Weight']].plot(kind='box', subplots=True, layout=(2,4), sharex=False, sharey=False, figsize=(20,10))
plt.show()
```

```
import seaborn as sns
correlation = data.corr()
plt.figure(figsize=(10, 10))
sns.heatmap(correlation, vmax=1, square=True, annot=True, cmap='cubehelix')
plt.title('Correlation between different features')
```

```
plt.figure(figsize=(20, 5))
features = ['Length', 'Diameter', 'Height', 'Weight', 'Shucked Weight', 'Viscera Weight', 'Shell Weight']
for i, feature in enumerate(features):
    plt.subplot(1, 7, i+1)
    data.plot(kind='scatter', x=feature, y='Age', ax=plt.gca())
    plt.xlabel(feature)
    plt.ylabel('Age')
plt.tight_layout()
```

```
plt.figure(figsize=(10, 5))
sns.histplot(data['Age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
```
